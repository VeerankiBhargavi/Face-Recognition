```python
import cv2
import os
import time

def detect_faces(image_path):
    # Load the pre-trained Haar Cascade classifier for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    # Read the image
    image = cv2.imread(image_path)
    
    # Convert the image to grayscale (required for face detection)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the grayscale image
    faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    # Draw rectangles around the detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    return image

def process_images(input_dir):
    # Process each image in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(input_dir, filename)
            print("Processing:", image_path)
            detected_image = detect_faces(image_path)
            cv2.imshow("Detected Faces", detected_image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
            print("Done processing:", image_path)

if __name__ == "__main__":
    # Specify the directory path where your images are located
    input_dir = './faceDetection_images'
    
    start_time = time.time()
    process_images(input_dir)
    end_time = time.time()
    print("Total processing time:", end_time - start_time, "seconds")
```
